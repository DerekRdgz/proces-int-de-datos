```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
ingreso = [1452,1361,1426,1470,1456,1430,1354,1442,1394,1459,
           1399,1458,1537,1425,1445,1439,1348,1450,1431,1446,
           1485,1405,1461,1490,1426]

habitaciones = [23,47,21,39,37,29,23,44,45,16,
                30,42,54,27,34,15,19,38,44,47,
                43,38,51,61,39]

df = pd.DataFrame({
    "Habitaciones": habitaciones,
    "Ingreso": ingreso
})

df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

m, b = np.polyfit(df["Habitaciones"], df["Ingreso"], 1)

x_linea = np.linspace(df["Habitaciones"].min(), df["Habitaciones"].max(), 100)
y_linea = m * x_linea + b

plt.figure()
plt.scatter(df["Habitaciones"], df["Ingreso"])
plt.plot(x_linea, y_linea)  # línea de tendencia
plt.xlabel("Habitaciones ocupadas")
plt.ylabel("Ingreso del restaurante")
plt.title("Correlación con línea de tendencia")
plt.show()

print("Pendiente (m):", m)
print("Intercepto (b):", b)
```

```python
correlacion, p_valor = pearsonr(df["Habitaciones"], df["Ingreso"])

print("Coeficiente de correlación (r):", correlacion)
print("Valor p:", p_valor)
```

### Análisis
La gráfica de dispersión muestra una tendencia positiva entre el número de habitaciones ocupadas y el ingreso del restaurante aunque se ve algo débil.
El coeficiente de correlación de Pearson fue r = 0.423.
El valor p = 0.035 es menor que 0.05, por lo tanto la relación es estadísticamente significativa.
Esto sugiere que existe evidencia suficiente para afirmar que a mayor ocupación del hotel, el ingreso del restaurante tiende a aumentar, aunque la relación no es fuerte.
