```python
import random
```

```python
class Individuo:
    def __init__(self, cromosoma: list):
        self.cromosoma = cromosoma
        self.aptitud = 0
```

```python
class Poblacion:
    def __init__(self, tamaño: int, longitud_cromosoma: int):
        self.generacion = 0
        self.individuos = [
            Individuo([random.randint(0, 1) for _ in range(longitud_cromosoma)])
            for _ in range(tamaño)
        ]
    
    def mejor(self):
        return max(self.individuos, key=lambda ind: ind.aptitud)
```

```python
def evaluar(individuo):
    individuo.aptitud = sum(individuo.cromosoma)
```

```python
pob = Poblacion(tamaño=5, longitud_cromosoma=6)

print("=== Población inicial ===")
for ind in pob.individuos:
    evaluar(ind)
    print(f"  Cromosoma: {ind.cromosoma}  Aptitud: {ind.aptitud}")

print(f"\nMejor individuo: {pob.mejor().cromosoma}  Aptitud: {pob.mejor().aptitud}")
```

## 1.1 La naturaleza como optimizadora

La naturaleza lleva millones de años perfeccionando a los seres vivos mediante
la evolución. Solo sobreviven los mejor adaptados y sus características se heredan
a las siguientes generaciones. Darwin llamó a este proceso selección natural.

Científicos como Rechenberg (1965) y Holland (1960s) se inspiraron en esto para
crear los Algoritmos Genéticos (AGs): métodos computacionales que imitan la
evolución para encontrar soluciones óptimas a problemas complejos.

## 1.2 Un poco de biología

Los seres vivos almacenan su información en cromosomas (cadenas de ADN).
Al reproducirse, los cromosomas del padre y la madre se cruzan generando
hijos con características de ambos (meiosis). Ocasionalmente la enzima que
copia el ADN comete errores: eso es una mutación. Casi siempre es mala,
pero a veces da ventaja al individuo.

Conceptos clave:
- **Genotipo**: información genética del individuo
- **Fenotipo**: características físicas observables
- **Gen / Alelo**: unidad de información / valor del gen
- **Mutación**: cambio aleatorio en el ADN
