# Regresión Lineal Múltiple - SalsBerry Realty
1. Determinar las variables independientes
2. Determinar la variable dependiente
3. Obtener el valor de intersección
4. Estimar el corto de calefacción para una Temperatura externa media de 30°, 5 pulgadas de aislamiento y un calentador con antigüedad de 10 años.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#importar datos
df = pd.read_csv('SalsberryRealty.csv')
df.head()
```

# 1. Determinar las variables independientes

x1, x2 y x3

```python
X = df[['Temp', 'Insul', 'Age']]
X
```

```python
# 2. Determinar las variables Dependientes
```

```python
y =df[['Cost']]
y
```

```python
#Ajustar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X, y) #entrenando el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas
mse = mean_squared_error(y, y_pred) #Error de la media cuadrática
r2 = r2_score(y, y_pred) #varianza
```

# 3. Calcular el valor de Intersección

```python
model.intercept_
```

```python
#coeficiente
print("Coeficientes: ", model.coef_)
print("Error cuadrático medio", mse)
print("Varianza: ", r2)
```

# 4. Estimar pasa 30°, aislamiento 5", antigüedad 10 años

```python
Y = model.predict([[30, 5, 10]])
print("El costo de calefacción estimado es:", Y[0])
```
