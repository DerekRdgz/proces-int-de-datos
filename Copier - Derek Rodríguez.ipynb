```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Librerias para opreaciones de regresi칩n Lineal
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("./CopierSalesofAmerica.csv")
```

```python
#Determinen la cantidad de filas y columnas del archivo
data.shape
```

```python
#Mostrar la tabla
data.head()
```

```python
#Seleccionar los datos para determinar el coeficiente de correlacion
X_train = np.array(data["llamadas"]).reshape(-1,1)
y_train = np.array(data["ventas"])
```

```python
#Calcular el coeficiente de correlaci칩n lineal
r = linear_model.LinearRegression()
#Entrenando el modelo con los datos
r.fit(X_train, y_train)

#Obtener las predicciones
#Valores para y
y_pred = r.predict(X_train)

print("Coeficiente de correlacion: ", r.coef_)
print("Interseccion: ", r.intercept_)
print("Error cuadr치tico de la media: ", mean_squared_error(y_train, y_pred))
print("Varianza: ", r2_score(y_train, y_pred))
```

```python
#Realizar la grafica de dispersi칩n y la pendiente
x = X_train.ravel()
plt.figure(figsize=(8,5))
plt.scatter(x, y_train, color="steelblue", s=80, label="Datos")
plt.plot(x, y_pred, color="red", linewidth=3, label="Pendiente")
plt.xlabel("No. de llamadas")
plt.ylabel("Copiadoras vendidas")
plt.title("Copier Sales of America")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
